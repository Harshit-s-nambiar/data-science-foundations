## Q1. A company claims that their new battery lasts longer than their previous model, which has a mean lifespan of 100 hours in previous model. A sample of 25 new batteries is tested, yielding a mean lifespan of 105 hours with a standard deviation of 10 hours. At a 5% significance level, can we conclude that the new battery lasts longer?Formulate the null and alternative hypotheses, choose an appropriate test, and conduct the test to reach a conclusion.


```python
from scipy import stats
from scipy.stats import ttest_1samp
import math
```

```python
sample_mean = 105
pop_mean = 100
s = 10
alpha=0.05
n=25
Ho="There is no significant differnce in the battery life mean"
H1="New battery last longer"
```

```python
se=s/math.sqrt(n)
se
```

```python
t_calc=(sample_mean-pop_mean)/se
t_calc
```

```python
df=n-1
df
```

```python
p_value= 1- stats.t.cdf(t_calc,df)
p_value
```

```python
if p_value<alpha:
    print("we reject null,ie new battery lasts longer")

else:
    print("we fail to reject null")
```

## Q7. Explain Bayes theorem in detail with some example

## answer: Bayesâ€™ Theorem is a formula used to find the probability of an event occurring based on prior knowledge or evidence related to that event.

## formula : p(A|B)=(P(A)*P(B|A))/P(B)
## where 
## P(A|B)- probability of A where B has already occured
## P(B|A)- probability of B where A has already occured
## P(A) - probability of A
## P(B) - probability of B


```python
## example:
## P(Rain)=0.3
## p(Clouds)=0.5
## p(Clouds|Rain)= 0.8
## then , P(Rain|Clouds)= (P(clouds|Rain)*P(Rain))/P(Clouds)
## then P(Rain|Clouds)= 
(0.8*0.3)/0.5
```
