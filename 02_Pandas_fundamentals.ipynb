## 3.Explain the concept of missing values. How can you identify the missing values in a Pandas DataFrame ? What are the different ways of treating/Imputing/Deleting the missing values.Explain with example.


```python
"""
CONCEPT OF MISSING VALUES:

Missing values are the data that is not stored or recorded for some reason.
In pandas, missing values are usually represented as:
(a)NaN (Not a Number) — used for numeric data.
(b)None — used for object (string) data, but internally often converted to NaN.


Causes of missing values:

(a)Data entry mistakes.
(b)Incomplete surveys or forms.
(c)Mismatched joins/merges between tables.
(d)Data corruption.
"""
```

```python
"""
IDENTIFING MISSING VALUES IN PANDAS:
"""
import numpy as np
import pandas as pd

df=pd.DataFrame({
    "Name": ["A", "B", None, "D"],
    "Age": [25, np.nan, 30, 22],
    "City": ["Delhi", None, "Mumbai", "Kochi"]
})
df
```

```python
#To find the missing values:
df.isnull()        # True where values are missing
df.notnull()       # True where values are present
df.isnull().sum()  # Count missing values per column
df.isnull().sum().sum()  # Total missing values in entire DataFrame
```

```python
"""
TREATING AND IMPORTING MISSING VALUES:

There are 3 different strategies:

A) Deleting Missing Values
When missing data is small and can be removed without losing important information.

"""
df.dropna()         # Remove rows with any missing value
df.dropna(axis=1)   # Remove entire columns with missing values
```

```python
"""
B) Imputing / Filling Missing Values
When you can replace missing values with some meaningful substitute.

"""

df.fillna(0)  # Fill with a constant

df["Age"].fillna(df["Age"].mean(), inplace=True) # Fill with column mean (numeric data)

df["Age"].fillna(df["Age"].median(), inplace=True) # Fill with column median (numeric data)

df["City"].fillna(df["City"].mode()[0], inplace=True) # Fill with mode (most frequent value, works for strings too)

df.fillna(method="ffill") # Forward fill (propagate last valid value forward)

df.fillna(method="bfill") # Backward fill (propagate next valid value backward)
```
