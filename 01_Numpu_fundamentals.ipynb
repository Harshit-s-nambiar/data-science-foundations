## Q4. Create a 2-D numpy array of random integers between 1 to 100 with shape 5*4. Now perform following operations:
## • Print only the even numbers from this array
## • Print only the odd numbers from this array
## • Print row wise sum of all elements
## • Print column wise sum of all elements
## • Convert this 2-D array into 3-D array of shape 2*2*5


```python
a=np.random.randint(1,100,20).reshape(5,4)
a
```

```python
# Print only the even numbers from this array
print("even numbers :",a[np.where(a%2==0)])
```

```python
# Print only the odd numbers from this array
print("odd numbers :",a[np.where(a%2!=0)])
```

```python
# Print row wise sum of all elements
print("row-wise sum of elements :",np.sum(a,axis=1))
```

```python
#  Print column wise sum of all elements
print("column-wise sum of elements :",np.sum(a,axis=0))
```

```python
# Convert this 2-D array into 3-D array of shape 2*2*5
print(a.reshape(2,2,5))
```

## Q5. Create a 2-D numpy array of random integers between 1 to 100 with shape 5*4. Now perform following operations:
## • Swap first row with third row
## • Swap second column with fourth column.
## • Replace the values less than 50 with zero.
## • Convert this 2D array into 1-D array


```python
b=np.random.seed(23)
b=np.random.randint(1,100,20).reshape(5,4)
print("b before swapping:\n",b)
```

```python
# Swap first row with third row:

b[[0,2]]=b[[2,0]]
print("b after swapping first row with third row:\n",b)
```

```python
# Swap second column with fourth column:
b[:,[1,3]]=b[:,[3,1]]
print("after swapping 2nd column with 4th column\n",b)
```

```python
# Replace the values less than 50 with zero:
c=np.where(b<50,0,b)
c
```

```python
# Convert this 2D array into 1-D array:
d=c.flatten()
d
```
