```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

## 1. Import the attached Indian_food.csv file in Jupyter notebook and perform following data cleaning operations through operations using Pandas and visualizations using Matplotlib, Seaborn and Plotly:



```python
df=pd.read_csv("Indian_food.csv")
df.head()
```

## a)In the state column, replace the -1 with the mode of state column.

```python
state_mode=df["state"].mode()[0]
state_mode
```

```python
df["state"]=df["state"].replace("-1",state_mode)
df.head(15)
```

## b) Whatever the state you got in previous question, fill the corresponding region in the region column.

```python
df.loc[df["state"]=="Gujarat","region"]="West"
df.head(15)
```

## c) Replace the -1 with mean for cook time variable

```python
df["cook_time"].unique()
```

```python
cook_mean=df["cook_time"].mean()
cook_mean
```

```python
df["cook_time"]=df["cook_time"].replace(-1,cook_mean)
df.tail()
```

## After performing these data cleaning steps, perform the below visualizations. You can use any of the 3 libraries (Matplotlib, seaborn, Plotly): 

## d)Most preferred flavours by number of customers using pie chart.


```python
flv=df["flavor_profile"].value_counts()
plt.pie(flv,labels=flv.index,autopct="%1.1f%%")
plt.title("Most preferred flavours by number of customers .")
plt.show()
```

## e) Most requested course by customers using Bar chart. Print the labels also on the top of each bar. Then sort the bar chart in descending order.

```python
course_mode=df["course"].value_counts().sort_values(ascending=True)
plt.bar(course_mode.index,course_mode.values)
for x,y in zip(course_mode.index,course_mode.values):
    plt.text(x,y,str(y),ha="center",va="bottom")

plt.xlabel("Course")
plt.ylabel("no.of customers")
plt.title("Most requested course by customers")
plt.show()
```

## f) Comparison between Vegetarians and Non-Veg orders in each region. Hint: You can use stacked/grouped bar chart for comparison. 

```python
diet_chart=df.groupby(["region","Diet"]).size().reset_index(name="count")
px.bar(diet_chart,x="region",y="count",barmode="group",color="Diet",title="Comparison between Vegetarians and Non-Veg orders in each region.")
```

## g) Create the box plot for Order cooking time to see the distribution. Also check if there are any outliers in this column.

```python

sns.boxplot(df,y="cook_time",x="course")
plt.title("Distribution for order cooking time")
plt.show()
# This graph include every outliers 
```

```python

sns.boxplot(df,y="cook_time",x="course")
plt.ylim(0,100)
plt.title("Distribution for order cooking time")
plt.show()
# THis is the same above graph but has a zoomed view of the box plot distribution.
```
